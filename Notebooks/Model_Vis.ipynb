```python
from io import StringIO
import json

yelp = []
texts = []
for line in StringIO( data.decode("utf-8")):
    j = json.loads(line)
    yelp.append( j)
    texts.append( j['text'])
#print(yelp[0]['text'])
print( len(texts), texts[0])
```

```python
textTokens = [[word for word in gensim.utils.tokenize(text, lowercase=True, deacc=True, errors="ignore")
                     if word not in stoplist]
              for text in texts]
dict3 = gensim.corpora.Dictionary( documents=textTokens)
textBOW = [dict3.doc2bow(text) for text in textTokens]
model2 = gensim.models.ldamodel.LdaModel( corpus=textBOW, id2word=dict3,
                                          num_topics=20, update_every=0.5,
#                                          iterations=100, passes=3)
                                          iterations=10, passes=1) # ONLY FOR FASTER TESTING
```

```python
pyLDAvis.enable_notebook()
data = pyLDAvis.gensim.prepare( model2, textBOW, dict3)
```
